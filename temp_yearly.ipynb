```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dask.dataframe as ddf
from pandas import Series, DataFrame
%matplotlib inline
```

```python
import seaborn as sn
import plotly.express as px
```

```python
years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
        '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019',
         '2020']
```

```python
ur_files = ddf.read_csv('daily_temp_county_level/daily_temp_county_level_2000.csv', dtype = {'fips': str})
temp = ur_files.compute()
temp = temp.loc[:, ~temp.columns.str.contains('^Unnamed')]
temp = temp.groupby('fips').mean()
temp = temp.reset_index()
temp = temp.drop('year', axis = 1)
temp['year'] = '2010'
temp.head()
```

```python
for year in years:
    ur_files = ddf.read_csv('daily_temp_county_level/daily_temp_county_level_' + year + '.csv', dtype = {'fips': str})
    temp = ur_files.compute()
    temp = temp.loc[:, ~temp.columns.str.contains('^Unnamed')]
    temp = temp.groupby('fips').mean()
    temp = temp.reset_index()
    temp = temp.drop('year', axis = 1)
    temp['year'] = year
    temp.to_csv('temp_average_yearly/temp_average_yearly_' + year + '.csv')
```

### 3 years - average temperature

```python
year_list = [['2000', '2001', '2002'], ['2003', '2004', '2005'], ['2006', '2007', '2008'],
            ['2009', '2010', '2011'], ['2012', '2013', '2014'], ['2015', '2016', '2017'],
            ['2018', '2019', '2020']]
```

```python
year_last = ['2000 to 2002', '2003 to 2005', '2006 to 2008', '2009 to 2011', 
             '2012 to 2014', '2015 to 2017', '2018 to 2020']
```

```python
for years,item in zip(year_list, year_last):
    frames = []
    for year in years:
        ur_files = ddf.read_csv('temp_average_yearly/temp_average_yearly_' + year + '.csv', dtype = {'fips': str})
        temp = ur_files.compute()
        temp = temp.loc[:, ~temp.columns.str.contains('^Unnamed')]
        frames.append(temp)
    temp_years = pd.concat(frames)
    temp_years = temp_years.groupby('fips').mean()
    temp_years = temp_years.reset_index()
    temp_years['period'] = item
    temp_years = temp_years.drop('year', axis = 1)
    temp_years.to_csv('temp_average_3years/temp_average_' + item + '.csv')
```

```python
ur_files = ddf.read_csv('temp_average_3years/temp_average_2000 to 2002.csv', dtype = {'fips': str})
temp = ur_files.compute()
temp = temp.loc[:, ~temp.columns.str.contains('^Unnamed')]
temp.head()
```

### 3 years - days of temperature in different intervals

```python
for years,item in zip(year_list, year_last):
    frames = []
    for year in years:
        ur_files = ddf.read_csv('Temp_Count_New/temp_count_county_level_' + year + '.csv', dtype = {'fips': str})
        temp = ur_files.compute()
        temp = temp.loc[:, ~temp.columns.str.contains('^Unnamed')]
        frames.append(temp)
    temp_count = pd.concat(frames)
    temp_count = temp_count.groupby('fips').sum()
    temp_count = temp_count.reset_index()
    
    temp_count['period'] = item
    temp_count = temp_count.drop('year', axis = 1)
    temp_count.to_csv('Temp_Count_3years/temp_count_' + item + '.csv')
```
